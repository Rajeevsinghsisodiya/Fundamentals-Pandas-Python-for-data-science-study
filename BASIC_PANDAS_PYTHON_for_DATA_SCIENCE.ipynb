1. Creating a Series

   A Series can be created from a list, dictionary, or scalar value

```python
import pandas as pd
years = [y for y in range(1995, 2001)]
series = pd.Series(years)
print(series)
```

2. Creating a DataFrame

```python
import pandas as pd
# Reading a CSV file into a DataFrame, not an excel file
df = pd.read_csv("/content/star_classification.csv")
print(df.head())
```

```python
import pandas as pd
from sqlite3 import connect
# Executing a SQL quering and returning a DataFrame df = pd.read_sql("SELECT * FROM people", conn)

# Establish connection
conn = connect(':memory:')

# Read CSV file into a pandas DataFrame
df = pd.read_csv("/content/star_classification.csv")

# Write the DataFrame to a SQLite table
df.to_sql("star_classification", conn, if_exists='replace', index=False)

# Execute SQL query on the newly created table
df_from_sql = pd.read_sql("SELECT * FROM star_classification", conn)

print(df_from_sql.head())
conn.close() # Close the connection
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for demonstration (replace 'people' with your actual data)
people = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 27, 22, 32, 26],
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles']
}

# Creating a DataFrame from the data
df = pd.DataFrame(people)

# Creating a bar plot for the 'City' column
# 'value_counts()' counts the occurrences of each city,
# 'plot(kind='bar')' generates
# Bar plot
df['Name'].value_counts().plot(kind='bar')
plt.show()

```
